```python
import pandas as pd
import numpy as np
# from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn import metrics
from sklearn import model_selection
from IPython.display import Image
import pydotplus
import os 
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz 2.44.1/bin/'
```

```python
data = pd.read_csv('C:/Workspace/Comento_Data_Analyst/train_mdf.csv')
```

```python
data = data[['MSSubClass','OverallQual','SalePrice']]
data
```

```python
# SalePrice를 범주화

for i in data['SalePrice']:
    if i <= 100000:
        data = data.replace(i, 0)
    elif 100000 < i <= 200000:
        data = data.replace(i, 1)
    elif 200000 < i <= 300000:
        data = data.replace(i, 2)
    elif 300000 < i <= 400000:
        data = data.replace(i, 3)
    elif 400000 < i <= 500000:
        data = data.replace(i, 4)
    elif 500000 < i <= 600000:
        data = data.replace(i, 5)
    elif 600000 < i <= 700000:
        data = data.replace(i, 6)
        
data
```

```python
X = np.array(pd.DataFrame(data, columns=['OverallQual','MSSubClass']))
y = np.array(pd.DataFrame(data, columns=['SalePrice']))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=None, max_leaf_nodes=None,
                                   max_features=None)     
clf = clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_train) 
score = metrics.accuracy_score(y_train, y_predict)
print(score) #1.0

y_predict = clf.predict(X_test) 
score = metrics.accuracy_score(y_test, y_predict)
print(score) #1.0
```

```python
feature_names = data.columns.tolist()
feature_names = feature_names[0:2]
target_name = np.array(['A','B','C','D','E','F','G']) #종속변수가 범주형이여야함
```

```python
dt_dot_data = export_graphviz(clf, out_file = None,
                                  feature_names = feature_names,
                                  class_names = target_name,
                                  filled = True, rounded = True,
                                  special_characters = True)
```

```python
dt_graph = pydotplus.graph_from_dot_data(dt_dot_data)
```
